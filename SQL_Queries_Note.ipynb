Notebook này dùng để lưu trữ và chạy các câu lệnh SQL phục vụ cho các mục đích khác nhau (ví dụ: Scheduling, Monitoring...).

### 1. Cấu hình kết nối
Nhập module `db_connection` để dễ dàng chuyển đổi giữa các database test.

```python
import db_connection
import pandas as pd

def run_query(db_num, query):
    conn = db_connection.get_connection(db_num)
    try:
        # Sử dụng pandas để hiển thị kết quả đẹp hơn dưới dạng bảng
        df = pd.read_sql_query(query, conn)
        return df
    finally:
        conn.close()
```

### 2. Truy vấn dữ liệu cho Task Schedule
Dưới đây là một số câu lệnh SELECT mẫu để kiểm tra trạng thái các task cần lập lịch.

#### Câu query 1: Lấy danh sách các task đang ở trạng thái 'Pending'

```python
query_pending = "SELECT * FROM tasks WHERE status = 'Pending'"

print("--- Kết quả từ Database 1 ---")
display(run_query(1, query_pending))

print("\n--- Kết quả từ Database 2 ---")
display(run_query(2, query_pending))
```

#### Câu query 2: Lấy các task có hạn chót (due_date) vào ngày hôm nay hoặc sắp tới

```python
query_due = "SELECT * FROM tasks WHERE due_date >= DATE('now') ORDER BY due_date ASC"

print("--- Task sắp tới từ Database 1 ---")
display(run_query(1, query_due))
```
