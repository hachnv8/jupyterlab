```python
import os
import pyodbc
from db_connections import CONNECTIONS, DRIVER_NAME

# Kiểm tra biến môi trường
print(f"Current User: {os.environ.get('USERNAME')}")
print(f"Current Path: {os.environ.get('PATH')[:100]}...")

# debug chuỗi kết nối
cfg = CONNECTIONS["atv400"]
conn_str = (
    f"DRIVER={{{DRIVER_NAME}}};"
    f"SYSTEM={cfg['system']};"
    f"UID={cfg['user'].strip()};"
    f"PWD={cfg['password'].strip()};"
    "Prompt=0;"
    "SIGNON=4;"
)

print(f"\nConnection String (Che password):")
print(conn_str.replace(cfg['password'], '********'))

try:
    conn = pyodbc.connect(conn_str)
    print("\nSUCCESS!")
    conn.close()
except Exception as e:
    print("\nFAILED AGAIN:")
    print(e)
```
